# Model for Distribution

```python
df.columns
```

```python
X = df[["Population","Daily_Demand_MLD","Potable_Available_MLD",
        "NonPotable_Available_MLD","Recycled_Usage_MLD",
        "Rainwater_Harvest_MLD","Groundwater_Level_m","Water_Loss_%"]]


df["Distributed_Potable"] = df[["Daily_Demand_MLD", "Potable_Available_MLD"]].min(axis=1)


df["Total_NonPotable_Available"] = (df["NonPotable_Available_MLD"] +
                                    df["Recycled_Usage_MLD"] +
                                    df["Rainwater_Harvest_MLD"])


df["Distributed_NonPotable"] = df[["Daily_Demand_MLD", "Total_NonPotable_Available"]].min(axis=1)


y_potable = df["Distributed_Potable"]
y_nonpotable = df["Distributed_NonPotable"]
```

```python
df
```

```python
# Potable distribution
X_train, X_test, y_train, y_test = train_test_split(X, y_potable, test_size=0.2, random_state=42)
model_potable_dist = RandomForestRegressor(n_estimators=100, random_state=42)
model_potable_dist.fit(X_train, y_train)
y_pred_potable = model_potable_dist.predict(X_test)
```

```python
print("Potable Distribution RÂ²:", r2_score(y_test, y_pred_potable))
print("Potable Distribution RMSE:", mean_squared_error(y_test, y_pred_potable, squared=False))
```

```python

# Non-potable distribution
X_train, X_test, y_train, y_test = train_test_split(X, y_nonpotable, test_size=0.2, random_state=42)
model_nonpotable_dist = RandomForestRegressor(n_estimators=100, random_state=42)
model_nonpotable_dist.fit(X_train, y_train)
y_pred_nonpotable = model_nonpotable_dist.predict(X_test)
```

```python
print("Non-Potable Distribution RÂ²:", r2_score(y_test, y_pred_nonpotable))
print("Non-Potable Distribution RMSE:", mean_squared_error(y_test, y_pred_nonpotable, squared=False))
```

```python

plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred_potable, alpha=0.6, color="blue")
plt.plot([0,max(y_test)], [0,max(y_test)], "r--")  # Perfect prediction line
plt.xlabel("Actual Distributed Potable (MLD)")
plt.ylabel("Predicted Distributed Potable (MLD)")
plt.title("Potable Water Distribution Prediction")
plt.show()

plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred_nonpotable, alpha=0.6, color="green")
plt.plot([0,max(y_test)], [0,max(y_test)], "r--")
plt.xlabel("Actual Distributed Non-Potable (MLD)")
plt.ylabel("Predicted Distributed Non-Potable (MLD)")
plt.title("Non-Potable Water Distribution Prediction")
plt.show()
```

```python


feat_imp_potable = pd.Series(model_potable_dist.feature_importances_, index=X.columns).sort_values(ascending=False)
feat_imp_nonpotable = pd.Series(model_nonpotable_dist.feature_importances_, index=X.columns).sort_values(ascending=False)

feat_imp_potable.plot(kind="bar", color="blue", title="Potable Distribution Feature Importance")
plt.show()

feat_imp_nonpotable.plot(kind="bar", color="green", title="Non-Potable Distribution Feature Importance")
plt.show()
```

```python
df["Total_Distribution"] = df["Distributed_Potable"] + df["Distributed_NonPotable"]
```

```python
df.head()
```

```python

X = df[["Population","Daily_Demand_MLD","Potable_Available_MLD",
        "NonPotable_Available_MLD","Recycled_Usage_MLD",
        "Rainwater_Harvest_MLD","Groundwater_Level_m","Water_Loss_%"]]

y_total = df["Total_Distribution"]


X_train, X_test, y_train, y_test = train_test_split(X, y_total, test_size=0.2, random_state=42)


model_total_dist = RandomForestRegressor(n_estimators=100, random_state=42)
model_total_dist.fit(X_train, y_train)

# Predictions
y_pred_total = model_total_dist.predict(X_test)

# Evaluation
print("Total Distribution RÂ²:", r2_score(y_test, y_pred_total))
print("Total Distribution RMSE:", mean_squared_error(y_test, y_pred_total, squared=False))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred_total, alpha=0.6, color="purple")
plt.plot([0, max(y_test)], [0, max(y_test)], "r--")
plt.xlabel("Actual Total Distribution (MLD)")
plt.ylabel("Predicted Total Distribution (MLD)")
plt.title("Total Water Distribution Prediction")
plt.show()
```
